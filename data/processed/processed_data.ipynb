```python
import numpy as np
import pandas as pd

df = pd.read_csv('spotify-tracks-dataset.csv')
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0.1','Unnamed: 0'])
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['track_id'] = df['track_id'].astype('string')
df['artists'] = df['artists'].astype('string')
df['album_name'] = df['album_name'].astype('string')
df['track_name'] = df['track_name'].astype('string')
df['track_genre'] = df['track_genre'].astype('string')

df = df.dropna()

df.info()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates(subset=['track_name', 'artists'], keep='first')
```

```python
df.duplicated().sum()
```

```python
df = df.drop(columns=['track_id','artists','album_name','track_name'])
df.head()
```

```python
df['track_genre'].nunique()
```

```python
df['explicit'] = df['explicit'].astype(int)
```

```python
df_cleaned = df
df_cleaned.head(20)
```

```python
df.info()
```

```python
def mood_label(df):
    valence = df['valence'].median()
    energy = df['energy'].median()
    print(f"Threshold Valence: {valence:.4f}")
    print(f"Threshold Energy: {energy:.4f}\n")

mood_label(df_cleaned)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sample_df = df_cleaned.sample(n=1000, random_state=42)

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 8))

sns.scatterplot(
    data=sample_df,
    x='valence',
    y='energy',
    alpha=0.3,   
    s=20,        
    color='teal',
    edgecolor=None)
plt.axvline(0.5, color='red', linestyle='--', linewidth=1.5, alpha=0.7, label='Decision boundary 0.5')
plt.axhline(0.5, color='red', linestyle='--', linewidth=1.5, alpha=0.7)

plt.title('Valence vs Energy Distribution', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Valence', fontsize=12)
plt.ylabel('Energy', fontsize=12)
plt.legend()
plt.tight_layout()

plt.savefig('raw_2d_scatter.png')
plt.close()
```
