## Baseline Model
Applied algorithm: Multinomial Logistic Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    classification_report,
)

df = pd.read_csv("cleaned_data.csv")
df.head()
```

```python
# Define a threshold for classifying the four moods based on valence and energy
def assign_mood(row):
    if row['valence'] >= 0.5 and row['energy'] >= 0.5:
        return 'Happy'
    elif row['valence'] < 0.5 and row['energy'] >= 0.5:
        return 'Angry'
    elif row['valence'] >= 0.5 and row['energy'] < 0.5:
        return 'Calm'
    else:
        return 'Sad'

# Remove "valence" and "energy" from the list of features to avoid data leakage

df['mood'] = df.apply(assign_mood, axis=1)
features = ['popularity', 'duration_ms', 'explicit', 'danceability', 
            'key', 'loudness', 'mode', 'speechiness', 'acousticness', 
            'instrumentalness', 'liveness', 'tempo', 'time_signature']
numerical_cols = ['popularity', 'duration_ms', 'explicit', 'danceability',
                      'loudness', 'speechiness', 'acousticness', 
                      'instrumentalness', 'liveness', 'tempo']
categorical_cols = ['key', 'mode', 'time_signature']
X = df[features]
y = df["mood"]
```

Spliting into train/dev/test with proportion 80/10/10

```python
# Split train/test
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.10, random_state=42, stratify=y
)

# Split train/dev
X_train, X_dev, y_train, y_dev = train_test_split(
    X_temp, y_temp, test_size=0.111, random_state=42, stratify=y_temp
)

print(f"Train: {len(X_train)} | Dev: {len(X_dev)} | Test: {len(X_test)}")
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ]
)

log_reg = Pipeline(
    [
        ('preprocessor', preprocessor),
        ('model', LogisticRegression(solver='lbfgs', max_iter=1000))
    ]
)

log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_dev)
print(y_pred[:5])
y_prob = log_reg.predict_proba(X_dev)
res = pd.DataFrame({"true_mood": y_dev.values, "predicted_mood": y_pred, "predicted_probability": y_prob.max(axis=1)})
res.head(5)
```

Note: The sum of predicted probability for each of the 4 class in each song = 1

```python
accuracy = accuracy_score(y_dev, y_pred)
precision = precision_score(y_dev, y_pred, average="macro")
recall = recall_score(y_dev, y_pred, average="macro")
f1 = f1_score(y_dev, y_pred, average="macro")

print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1 Score : {f1:.4f}")
```

```python
cm = confusion_matrix(y_dev, y_pred)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=log_reg.classes_
)

disp.plot(cmap='Blues')
plt.title("Baseline Logistic Regression (Dev)")
plt.tight_layout()
plt.show()
```

Problems:
1. Angry - happy mistakes is large (with 907 angry predicted as happy and 725 happy predicted as angry)
2. Recall on calm is too low

```python
print("Train: ", accuracy_score(y_train, log_reg.predict(X_train)))
print("Dev: ", accuracy_score(y_dev, log_reg.predict(X_dev)))
```

Problem: Underfitting

Solution: Since valence and energy where the 2 features that we based on to separate the moods, but they were removed,
we can add a new feature that is more informative about mood

## Improved Model
1. Add a new feature about rythm intensity, which could suggest more about the song's arousal (more joy/aggressive or
calm/sadness).
2. Use decision tree to create non-linear decision boundaries to better separate the 4 moods.

```python
# Normalize tempo to 0-1 range
df['tempo_norm'] = (df['tempo'] - df['tempo'].min()) / (df['tempo'].max() - df['tempo'].min())

# Normalize loudness to 0-1 range
df['loudness_norm'] = (df['loudness'] - df['loudness'].min()) / (df['loudness'].max() - df['loudness'].min())

# Engineered feature
df['rhythm_intensity'] = (
    df['danceability'] * df['tempo_norm'] * (1 - df['acousticness'])
)
features = ['popularity', 'duration_ms', 'explicit', 'danceability', 
            'key', 'loudness', 'mode', 'speechiness', 'acousticness', 
            'instrumentalness', 'liveness', 'tempo', 'time_signature', 'rhythm_intensity']

numerical_cols = ['popularity', 'duration_ms', 'explicit', 'danceability',
                      'loudness', 'speechiness', 'acousticness', 
                      'instrumentalness', 'liveness', 'tempo', 'rhythm_intensity']

categorical_cols = ['key', 'mode', 'time_signature']

X = df[features]
y = df["mood"]

X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.10, random_state=42, stratify=y
)
X_train, X_dev, y_train, y_dev = train_test_split(
    X_temp, y_temp, test_size=0.111, random_state=42, stratify=y_temp
)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ]
)

log_reg = Pipeline(
    [
        ('preprocessor', preprocessor),
        ('model', LogisticRegression(solver='lbfgs', max_iter=1000))
    ]
)

log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_dev)
print(y_pred[:5])
y_prob = log_reg.predict_proba(X_dev)
res = pd.DataFrame({"true_mood": y_dev.values, "predicted_mood": y_pred, "predicted_probability": y_prob.max(axis=1)})
res.head(5)
```

```python
accuracy = accuracy_score(y_dev, y_pred)
precision = precision_score(y_dev, y_pred, average="macro")
recall = recall_score(y_dev, y_pred, average="macro")
f1 = f1_score(y_dev, y_pred, average="macro")

print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1 Score : {f1:.4f}")
```

```python
print("Train: ", accuracy_score(y_train, log_reg.predict(X_train)))
print("Dev: ", accuracy_score(y_dev, log_reg.predict(X_dev)))
```

Problem: added engineered feature does not improve the performance of the model

Solution: use stronger model - Decision Tree that helps create non-linear decision boundary (where the model usually make
more mistakes

```python
from sklearn.ensemble import RandomForestClassifier

for depth in [5, 10, 15, 20, 25, 30]:
    rf_temp = Pipeline([
        ('preprocessor', preprocessor),
        ('model', RandomForestClassifier(
            n_estimators=100, max_depth=depth, random_state=42, n_jobs=-1
        ))
    ])
    rf_temp.fit(X_train, y_train)
    train_acc = accuracy_score(y_train, rf_temp.predict(X_train))
    dev_acc   = accuracy_score(y_dev,   rf_temp.predict(X_dev))
    print(f"max_depth={depth:2d} | Train: {train_acc:.4f} | Dev: {dev_acc:.4f} | Gap: {train_acc - dev_acc:.4f}")
```

```python
rf = Pipeline([
    ('preprocessor', preprocessor),
    ('model', RandomForestClassifier(
        n_estimators=100,
        max_depth=15,       
        min_samples_leaf=5, 
        random_state=42,
        n_jobs=-1
    ))
])

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_dev)
accuracy = accuracy_score(y_dev, y_pred_rf)
precision = precision_score(y_dev, y_pred_rf, average="macro")
recall = recall_score(y_dev, y_pred_rf, average="macro")
f1 = f1_score(y_dev, y_pred_rf, average="macro")

print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1 Score : {f1:.4f}")
```

```python
print("Train:", accuracy_score(y_train, rf.predict(X_train)))
print("Dev  :", accuracy_score(y_dev,   rf.predict(X_dev)))
```

```python
# Final: Evaluation on test data
# With baseline model
y_pred_logreg_test = log_reg.predict(X_test)

acc_lr  = accuracy_score(y_test, y_pred_logreg_test)
prec_lr = precision_score(y_test, y_pred_logreg_test, average='macro')
rec_lr  = recall_score(y_test, y_pred_logreg_test, average='macro')
f1_lr   = f1_score(y_test, y_pred_logreg_test, average='macro')

# With improved model
rf_final = Pipeline([
    ('preprocessor', preprocessor),
    ('model', RandomForestClassifier(
        n_estimators=100, max_depth=15, random_state=42, n_jobs=-1
    ))
])
rf_final.fit(X_train, y_train)
y_pred_rf_test = rf_final.predict(X_test)

acc_rf  = accuracy_score(y_test, y_pred_rf_test)
prec_rf = precision_score(y_test, y_pred_rf_test, average='macro')
rec_rf  = recall_score(y_test, y_pred_rf_test, average='macro')
f1_rf   = f1_score(y_test, y_pred_rf_test, average='macro')

print("===BASELINE MODEL===")
print(f"Accuracy: {acc_lr:.4f}")
print(f"Precision: {prec_lr:.4f}")
print(f"Recall: {rec_lr:.4f}")
print(f"F1 score: {f1_lr:.4f}")
print("===BASELINE MODEL===")
print(f"Accuracy: {acc_rf:.4f}")
print(f"Precision: {prec_rf:.4f}")
print(f"Recall: {rec_rf:.4f}")
print(f"F1 score: {f1_rf:.4f}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

ConfusionMatrixDisplay(
    confusion_matrix(y_test, y_pred_logreg_test),
    display_labels=log_reg.classes_
).plot(cmap='Blues', ax=axes[0])
axes[0].set_title("Baseline - Logistic Regression (Test)")

ConfusionMatrixDisplay(
    confusion_matrix(y_test, y_pred_rf_test),
    display_labels=rf_final.classes_
).plot(cmap='Greens', ax=axes[1])
axes[1].set_title("Improved - Decision Tree (Test)")

plt.tight_layout()
plt.show()
```

## ERROR ANALYSIS

```python
# 1. Create an error analysis dataframe
analysis_df = X_test.copy()
analysis_df['true_mood'] = y_test
analysis_df['pred_mood'] = y_pred_rf_test
analysis_df['is_error'] = analysis_df['true_mood'] != analysis_df['pred_mood']

# 2. Pull valence and energy back from original data using the matching indexes
analysis_df['valence'] = df.loc[X_test.index, 'valence']
analysis_df['energy'] = df.loc[X_test.index, 'energy']
analysis_df['track_genre'] = df.loc[X_test.index, 'track_genre']

# 3. Get model confidence (prediction probabilities)
probabilities = rf_final.predict_proba(X_test)
analysis_df['pred_confidence'] = probabilities.max(axis=1)

total_errors = analysis_df['is_error'].sum()
```

## 1. The Boundary-Case Mistake

```python
margin = 0.05
boundary_errors = analysis_df[
    (analysis_df['is_error'] == True) & 
    ((abs(analysis_df['valence'] - 0.5) < margin) | (abs(analysis_df['energy'] - 0.5) < margin))
]
pct_boundary = (len(boundary_errors) / total_errors) * 100

print("1. BOUNDARY-CASE ANALYSIS:")
print(f"   -> Out of {total_errors} errors, {len(boundary_errors)} ({pct_boundary:.2f}%) "
      f"happen within {margin} of the 0.5 threshold line!")
print("   -> Interpretation: These tracks sound almost identical to the human ear; "
      "the error is an artifact of the hard 0.5 grid boundary.\n")
```

    The Explanation: 
        - A song with a valence of 0.49 and energy of 0.52 is strictly labeled "Angry". However, a song with a valence of 0.51 and energy of 0.52 is labeled "Happy". Mathematically, these two tracks are almost identical, and to a human ear, they sound exactly the same. The model may predict "Happy" for the first song. Even though the metric counts this as a "mistake," it isn't a catastrophic failure of the model; it is an artifact of forcing continuous, fluid human emotions into a rigid, hard-bounded box. 
        
    Analyzation Method: 
        - For all misclassified songs, calculate absolute distance to the threshold boundaries (|valence - 0.5|and |energy - 0.5|). If a vast majority of mistakes have a distance under 0.05,  model is actually performing reasonably well, but is being penalized by the hard border splits.

## 2. The Adjacent Leakage Mistake

    The Explanation:
        - Happy and Angry both share High Energy (>= 0.5).
        - Calm and Sad both share Low Energy (< 0.5)
        - If the model frequently misclassifies Sad songs as Calm, it means the model has successfully learned to identify "Low Energy" . However, it completely fails to distinguish the Valence (positivity vs. negativity). This indicates that the features intended to capture the emotional tone or brightness of the track are either weak or aren't being weighed heavily enough by the model.
        
    Analyzation Method: 
        - Look at the off-diagonal values of confusion matrix. Map out whether the "leakage" happens horizontally (confusion over Valence) or vertically (confusion over Energy). This will tell you exactly which musical dimension of feature set is failing to capture.

```python
print("2. DIRECTIONAL LEAKAGE ANALYSIS:")
just_errors = analysis_df[analysis_df['is_error'] == True]
leakage_matrix = pd.crosstab(just_errors['true_mood'], just_errors['pred_mood'])
print(leakage_matrix)
```

## 3. The High-Confidence, Blind-Spot Mistake

    The Explanation: 
    - This mistake occurs when the model outputs an incorrect prediction but does so with a very high probability score
    - This happens due to structural contradictions between standard acoustic features and the mathematical mood label. For instance, a heavy electronic or rock song (such as genres like techno, groove, or hard-rock) might have a mid-range tempo and low acoustic brightness (valence), which tricks the model into thinking the track belongs in a lower-intensity category like Angry or Sad. However, due to production styles, its true underlying energy parameter remains mathematically high. The model becomes "confidently blind" because it relies heavily on dominant tree splits (like acousticness or tempo) that send conflicting signals for these specific, highly processed genres.
    
    Analyzation Method:
    - To track down these blind spots, we filter our incorrect predictions using model.predict_proba() to isolate errors where the model's confidence exceeded 65%.

```python
confidence_threshold = 0.65
confident_blunders = analysis_df[
    (analysis_df['is_error'] == True) & 
    (analysis_df['pred_confidence'] >= confidence_threshold)
]

print(f"3. HIGH-CONFIDENCE BLUNDERS (Confidence >= {confidence_threshold*100}%):")
print(f"   -> The model made {len(confident_blunders)} highly confident but wrong decisions.")
print("   -> Samples of top misclassified tracks:")
print(confident_blunders[['track_genre', 'tempo', 'acousticness', 'true_mood', 'pred_mood', 'pred_confidence']].head(8))
```
